
### Multi-Agent LLM system for HR hiring decision support in game development


```python
# Multi-agent workforce planning and hiring approach for B2C product creation
```

```python
import autogen
```

```python
import os
```

```python
from getpass import getpass

os.environ['OPENAI_API_KEY'] = getpass('put OpenAI API key: ')
```

```python
# LLM Configuration for Agentic Multi-Agent System

openai_config = [{'model': 'gpt-4o-mini', 'api_key': os.environ['OPENAI_API_KEY'], 'temperature': 0.1}]

llm_config = {'config_list': openai_config}
```

```python
from autogen import ConversableAgent
```

```python
# Agentic HR Manager for multi-agent LLM collaboration
human_resources_agent = ConversableAgent(name='HR_Manager', 
                                         system_message='''
                                         You are the HR Manager responsible for staffing strategy. 
                                         The company is starting the development of a new B2C technology product and aims to build a specialized 
                                         project team. Your responsibility is to coordinate with different departments to assemble and synthesize
                                         their role requirements.''', 
                                         llm_config=llm_config,
                                         human_input_mode='NEVER')
```

```python
# Agentic product manager for multi-agentctional work
product_manager_agent = ConversableAgent(name ='Product_Manager',
                                         system_message='''
                                         You are the Head of Product at a company launching a new B2C technology product. 
                                         You need to coordinate with HR to outline the position requirements for the product management team. 
                                         Your responsibility is to provide a well-organized summary of the key roles, skills, credentials, and budget 
                                         needed for the project.
                                         ''',
                                         llm_config=llm_config, 
                                         human_input_mode='NEVER'
                                        )
```

```python
# Agentic engineering manager for multi-agent LLM collaboration

engineering_manager_agent = ConversableAgent(name = 'Engineering_Manager', 
                                             system_message = '''
                                             You are the head of engineering at a company launching a new B2C technology product. Your responsibility
                                             is to coordinate with the HR department to define the staffing requirements for the engineering team. 
                                             You need to give a well-structured summary of the key roles, required skills, qualifications, and budget 
                                             for the project.
                                             ''',
                                             llm_config=llm_config,
                                             human_input_mode = 'NEVER'
                                            )
```

```python
# Agentic marketing manager for multi-agent LLM coordination

marketing_manager_agent = ConversableAgent(name = 'Marketing_Manager', 
                                           system_message = ''' You are the Head of Marketing at a company launching a new B2C technology product.
                                           Your role is to collaborate with HR to identify the staffing requirements for the Marketing team. You need 
                                           to offer an organized summary of the primary roles, necessary skills, qualifications, and financial resources
                                           for the project.
                                           ''',
                                           llm_config=llm_config,
                                           human_input_mode="NEVER"
                                          )
```

```python
# Client support agent for multi-agent hiring system

support_manager_agent = ConversableAgent(name = 'Client_Support',
                                         system_message = '''
                                         You are the head of customer support at a company introducing a new B2C technology product. Your role is to 
                                         work with HR to determine the staffing requirements for a customer support team. You need to provide a 
                                         condensed overview of 
                                         - positions
                                         - essential skills
                                         - qualifications
                                         - budget planning. 
                                         ''',
                                         llm_config=llm_config,
                                         human_input_mode = 'NEVER'
                                        )
```

```python
from autogen import GroupChat
```

```python
# Agent coordination approach for workforce strategy 

team_coordination_chat = GroupChat(agents = [human_resources_agent,   
                               product_manager_agent,
                               engineering_manager_agent,
                               marketing_manager_agent,
                               support_manager_agent], messages = [], max_round = 4, speaker_selection_method = 'auto')
```

```python
# Group chat manager for multi-agent team alignment

from autogen import GroupChatManager

team_coordination_manager = GroupChatManager(groupchat = team_coordination_chat, 
                                             llm_config = llm_config)
```

### How do I say that the group chat is executing correctly?

```python
# Check for HR agent dialogue for team formation# Group chat manager for multi-agent team alignment

from autogen import GroupChatManager

team_coordination_manager = GroupChatManager(groupchat = team_coordination_chat, 
                                             llm_config = llm_config)
chat_result = human_resources_agent.initiate_chat(
    team_coordination_manager, message = 'The company is releasing a new  PS4 game in the mobile gaming domain. You need to find the staffing requirements and major positions.',
    summary_method = 'last_msg',)
```

```python
# Selected agent for message lookup
lookup_agent_name = support_manager_agent.name  # "Client_Support"

# Go through messages to get the last conversation of a selected agent (lookup_agent_name)

for chat_agent, dialogues in team_coordination_manager.chat_messages.items():
    if getattr(chat_agent, 'name', '') == lookup_agent_name:   # if true, go through
        if dialogues:         # Check if messages exist
            print(f'Most recent response from {lookup_agent_name}:\n')
            print(dialogues[-1]['content'])
        else:
            print(f' Agent {lookup_agent_name} has no chat history available.')
        break
    else:
        print(f'Agent {lookup_agent_name} has no chat messages.')
```

### Agentic HR workforce structuring with departmental summarization

```python
# Cross-department staffing summary agent

department_summary_agent = ConversableAgent(name = 'summary_manager',
                                    system_message = '''You are a Summary Agent in a game development studio. Your role is to review the hiring 
                                    requirements from HR, Product, Engineering, Marketing, and Client Support and then generate a systematic summary. 
                                    You need to list only the role titles under each department. You are not allowed to include salaries, skills, and 
                                    detailed descriptions in each department. 
                                    ''',
                                    llm_config = llm_config,
                                    human_input_mode = "NEVER",)
```

```python
# Group chat speaker routing

def get_next_speaker(previous_speaker, groupchat):
    '''
    Select the next agent in the interactive chat using the previous speaker.
    '''
    messages = groupchat.messages
    
    if previous_speaker is human_resources_agent:
        return product_manager_agent
    
    elif previous_speaker is product_manager_agent:
        return engineering_manager_agent
        
    elif previous_speaker is engineering_manager_agent:
        return marketing_manager_agent
    
    elif previous_speaker is marketing_manager_agent:
        return support_manager_agent
    
    elif previous_speaker is support_manager_agent:
        return department_summary_agent
    
    elif previous_speaker is department_summary_agent:
        return None
```

```python
# Full department staffing group chat with summary agent
department_chat = GroupChat(agents=[human_resources_agent,
                                    product_manager_agent,
                                    engineering_manager_agent,
                                    marketing_manager_agent,
                                    support_manager_agent,
                                    department_summary_agent],
                            messages=[],
                            max_round = 5,
                            speaker_selection_method="round_robin"
                           )
```

```python
department_chat_manager = GroupChatManager(groupchat = department_chat, llm_config = llm_config)
```

```python
# HR hiring overview chat
chat_response = human_resources_agent.initiate_chat(
    department_chat_manager,
    message = 'A company developing mobile games is releasing a new PS4 game. You need to determine the hiring needs for this product, listing positions and staffing needs.',
    summary_method = 'last_msg',
)
```

```python
# HR hiring overview chat
chat_response = human_resources_agent.initiate_chat(
    department_chat_manager,
    message = 'A toothpaste company is introducing a new mouthwash. You are responsible for compiling and summarizing the recruitment needs, including all essential roles for the project.',
    summary_method = 'last_msg',
)
```

### Digital marketing and creative team staffing coordination by multi-agent chat for  PS4 game release 

```python
# Digital Marketing Team hiring coordinator

digital_marketing_manager = ConversableAgent(
    'digital_marketing_manager',
    system_message = '''You are the Head of Digital Marketing at a game development company introducing a new PS4 game. You report directly to the 
    Marketing Manager.Your obligation is to coordinate with the management team to identify the hiring needs for structuring a digital marketing team 
    for this game launch.
    You are required to submit a comprehensive summary of:
    -  Required roles
    -  Qualifications
    -  Skills
    -  Projected budget
    ''',
    llm_config=llm_config, human_input_mode = 'NEVER',
)
```

```python
# Creative team hiring coordinator

creative_team_manager = ConversableAgent(
    name = 'creative_team_manager', 
    system_message = ''' You are the Head of the Creative Team at a game development company launching a new PS4 game. You need to report to the Head
    of Marketing. Your role is to coordinate with marketing department to define the hiring requirements for building a creative team for
    this project. You need to give a systematic summary of:
    -  the key roles
    -  required skills
    -  qualifications
    -  projected budget
    ''',
    llm_config = llm_config, human_input_mode = 'NEVER'
)
```

```python
# Nested marketing GroupChat initialization

marketing_group_chat = GroupChat(agents=[marketing_manager_agent,
                                         digital_marketing_manager,
                                         creative_team_manager],
                                 messages = [],
                                 max_round = 3,
                                 speaker_selection_method = 'round_robin'
                                )
```

```python
# Marketing GroupChat manager configuration
marketing_chat_manager = GroupChatManager(groupchat = marketing_group_chat)
```
